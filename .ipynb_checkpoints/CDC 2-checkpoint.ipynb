```python
import plt
import requests
```

### 사이트 접속 URL 입력

```python
url = 'http://www.cdc.go.kr/npt/biz/npp/iss/influenzaListAjax.do'
```

### 그냥 requests를 보내면 아무런 데이터가 오지않음 -> params 추가

> ### 연도별로 잘라진 걸 보고싶을 때 : 아래와 같은 파라미터를 사용(intoDivi 삭제)

```python
params = {
    'icdNm': 'influenza',
    'startYear': '2014',  # 몇 년 부터
    'endYear': '2020',  # 몇 년 까지
    'age': "",
    'ageText': '전체'  # 연령은 전체로
}
```

> ### 나이대 별로 잘라진 데이터를 보고 싶을 때 : 아래의 파라미터를 사용(intoDivi 포함)

```python
params = {
    'icdNm': 'influenza',
    'startYear': '2019',  # 몇 년 부터
    'endYear': '2020',  # 몇 년 까지
    'age': "",
    'intoDivi': 1,
    'ageText': '전체'  # 연령은 전체로
}
```

> ### 알맞은 파라미터를 골라서 requests를 날려준다

```python
response = requests.post(url, params)
response
```

> ### json 타입으로 response가 오기 때문에 json으로 읽기!
- `'intoDivi': 1` 을 파라미터에 추가할 경우 나이대 별로 데이터가 나온다
- `'intoDivi': 1` 를 파라미터에서 뺄 경우 연도별 데이터가 나온다

> #### 아래는 나이대별로 잘랐을 때의 예시이다

```python
datas = response.json()
```

```python
datas['data'][0]
```

```python
datas = datas['data']
```

## 나이대 별로 자를 경우 데이터 분석 예시

> ### 각 주차별 count가 VALUE에 저장되어 있으며, 나이대 별로 나누어져 있음

```python
datas
```

> ### 나이대 별로 자르기

```python
one = list(map(float, datas[0]['VALUE'].split("`")[:28]))
six = list(map(float, datas[1]['VALUE'].split("`")[:28]))
twleve = list(map(float, datas[2]['VALUE'].split("`")[:28]))
eightteen = list(map(float, datas[3]['VALUE'].split("`")[:28]))
fortynine = list(map(float, datas[4]['VALUE'].split("`")[:28]))
sixtyfour = list(map(float, datas[5]['VALUE'].split("`")[:28]))
after64 = list(map(float, datas[6]['VALUE'].split("`")[:28]))
```

> ### 그래프 그려보기

```python
x = range(1, 29)
plt.plot(x, one, label = '0세')
plt.plot(x, six, label = '1~6세')
plt.plot(x, twleve, label = '7~12세')
plt.plot(x, eightteen, label = '13~18세')
plt.plot(x, fortynine, label = '19~49세')
plt.plot(x, sixtyfour, label = '50~64세')
plt.plot(x, after64, label = '64세 이상')
plt.legend()
plt.show()
```
